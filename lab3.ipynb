```python
import pandas as pd
import math

df = pd.read_csv("student_data.csv")
print(df)
total = 14
yes = 0
no = 0
for row in df.values:
    if row[4] == "Yes":
        yes += 1
    elif row[4] == "No":
        no += 1

print("\nyes = " , yes)
print("no = " , no)

entropy_D = -(yes/total)*math.log2(yes/total)-(no/total)*math.log2(no/total)
print("Entropy(D) = ",entropy_D)
print(round(entropy_D,4))            

        
```

```python
p_yes = -(yes/total)*math.log2(yes/total)
print("p_yes =" , p_yes)
p_no = -(no/total)*math.log2(no/total)
print("p_no =" , p_no)
```
