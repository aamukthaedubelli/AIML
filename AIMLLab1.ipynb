```python
import numpy as np

arr = np.arange(1,21)
print("Array:")
print(arr)

print("\nShape of array:" , np.shape(arr))
print("\nShape of array:" , arr.shape)
print("\nMaximum value", arr.max())
print("\nMinimum value" , arr.min())
print("\nmean of the array" , arr.mean())
print("\n Standard deviation of the array" , arr.std())

matrix = arr.reshape(4,5)
print(matrix)
print("\nSecond row:")
print(matrix[1])

print("\nThird column:")
print(matrix[:,2])

      
```

```python
import pandas as pd
data = {
    "Roll No": [ 101 , 102 , 103 , 104 , 105],
    "Name" : ["Ravi" , "Priya" , "Kiran" , " Anu" , "Sita"],
    "Marks": [85, 92 , 78 , 88 , 98]
}
df = pd.DataFrame(data)
print("Data Frame:" ,df)
print("\nFirst Three Rows:" , df.head(3))
print("\nAverage marks:" , df["Marks"].mean())
print("\nStudent with Heighest Marks:\n" , df.loc[df["Marks"].idxmax()])
print("\nStudent with Lowest Marks:\n" , df.loc[df["Marks"].idxmin()])

def grade(Marks):
    if Marks >=90:
        return"A"
    elif Marks >=80:
        return "B"
    else:
        return "C"

df["Grade"] = df["Marks"].apply(grade)
print("\nDataFrame with Grade:\n" , df)
```

```python
import pandas as pd
data = {
    "Product": ["Laptop" , "Mouse"  ,"keyboard" , "Monitor"],
    "Price" : [50000 , 500 , 1200 , 10000],
    "Quantity": [10 , 50 , 30 , 15]
}
df = pd.DataFrame(data)
df["Total Value"] = df["Price"] * df["Quantity"]
print(df)
print("\nHeighest total value Product :\n" , df.loc[df["Total Value"].idxmax()])
print("\nOverall Inverntory Value:" , df["Total Value"] .sum())
```
